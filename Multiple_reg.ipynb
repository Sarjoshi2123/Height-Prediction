```python
#Creating a dataframe of this dataset
#Here using the california.data we are extracting all the independent features
df = pd.DataFrame(california.data,columns=california.feature_names)
df.head()
```

```python
#Now using the california.target we will add the dependent feture in our dataset
df['Price'] = california.target
```

```python
df.head()
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
#Now we will sparate out the independent and the dependent features
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
#Train test and split
from sklearn.model_selection import train_test_split
X_train, X_test, y__train, y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train.shape,y__train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
#Model training
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train, y__train)
```

```python
#Slope of coff
regression.coef_
```

```python
#intercept
print(regression.intercept_)
```

```python
#Predicted values
y_pred = regression.predict(X_test)
y_pred
```

```python
#Performance metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error, r2_score
print("mse :", mean_squared_error(y_test, y_pred))
print("mae :", mean_absolute_error(y_test, y_pred))
print("rmse :", root_mean_squared_error(y_test, y_pred))
print('r2 score :', r2_score(y_test, y_pred))
```

```python
#Assumptions
#scatter plot between the y-test and the y_pred
plt.scatter(y_test, y_pred)
```

```python
#Residuals
residuals = y_test- y_pred
sns.displot(residuals,kind='kde')
```

```python
#now a scatter plot between the residuals and the y_pred
plt.scatter(y_pred, residuals) #this should follow a linear pattern
```
